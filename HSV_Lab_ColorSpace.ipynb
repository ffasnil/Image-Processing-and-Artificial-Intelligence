```python
import argparse
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Set the file paths for the images
outdoor_path = r'C:\Users\ffah\Desktop\Projects\BEng Project\SIIT ML\Robotics Lecture\Robotics_code\week2\Figures\Cube1.jpg'
indoor_path = r'C:\Users\ffah\Desktop\Projects\BEng Project\SIIT ML\Robotics Lecture\Robotics_code\week2\Figures\Cube8.jpg'

# Read the images
Outdoor = cv2.imread(outdoor_path)
Indoor = cv2.imread(indoor_path)
```

```python
# HSV Color Space

# Convert images to HSV color space
OutdoorHSV = cv2.cvtColor(Outdoor, cv2.COLOR_BGR2HSV)
IndoorHSV = cv2.cvtColor(Indoor, cv2.COLOR_BGR2HSV)

# Display the original images using Matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(Outdoor, cv2.COLOR_BGR2RGB))
plt.title('Outdoor')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(Indoor, cv2.COLOR_BGR2RGB))
plt.title('Indoor')
plt.axis('off')

plt.show()

# Display the images in HSV color space
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

(Ho, So, Vo) = cv2.split(OutdoorHSV)
HSVspaceo = np.hstack([Ho, So, Vo])
ax1.imshow(HSVspaceo, cmap='gray')
ax1.set_title('HSV space: Outdoor')
ax1.axis('off')

(Hi, Si, Vi) = cv2.split(IndoorHSV)
HSVspacei = np.hstack([Hi, Si, Vi])
ax2.imshow(HSVspacei, cmap='gray')
ax2.set_title('HSV space: Indoor')
ax2.axis('off')

#-----------------------------------------------------
# Lab Color Space

# Convert images to LAB color space
OutdoorLab = cv2.cvtColor(Outdoor, cv2.COLOR_BGR2LAB)
IndoorLab = cv2.cvtColor(Indoor, cv2.COLOR_BGR2LAB)

# Unpack the LAB images into channels
(Lo, ao, bo) = cv2.split(OutdoorLab)
Labspaceo = np.hstack([Lo, ao, bo])

(Li, ai, bi) = cv2.split(IndoorLab)
Labspacei = np.hstack([Li, ai, bi])

# Display the LAB space images
fig, (ax3, ax4) = plt.subplots(1, 2, figsize=(15, 5))

ax3.imshow(Labspaceo, cmap='gray')
ax3.set_title('Lab space: Outdoor')
ax3.axis('off')

ax4.imshow(Labspacei, cmap='gray')
ax4.set_title('Lab space: Indoor')
ax4.axis('off')

plt.show()
```
