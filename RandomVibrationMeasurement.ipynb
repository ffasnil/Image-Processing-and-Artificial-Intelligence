```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Initialized 2 main parameters: mean and standard deviation
mean_displacement = 5 # unit: cm
stdev_displacement = 0.5 # unit: cm
# Generate random displacment measurement for 10,000 surveys
# In each survey, we take 10 measurements.
displacement_measurements = np.zeros((10000,10))
for i in range(displacement_measurements .shape[0]):
    for j in range(displacement_measurements .shape[1]):
        displacement_measurements [i,j] = mean_displacement + np.random.normal()*stdev_displacement

print('Mean displacement:', round(np.mean(displacement_measurements),1), 'cm')
print('Standard Deviation of displacment:', round(np.var(displacement_measurements)**0.5,1), 'cm')

# Histogram that shows the distribution for a single survey/epoch of measurements
fig, ax = plt.subplots(figsize=(10,6))
sns.histplot(displacement_measurements[30], bins=20, kde=False, label='Displacement')
ax.set_xlabel("Displacment [cm]",fontsize=16)
ax.set_ylabel("Frequency of occurance",fontsize=16)
plt.legend()
plt.tight_layout()
plt.savefig(fname='Distribution_EachEpoch', dpi=150)
plt.show()

# Histogram that shows the distribution for the mean of all surveys
fig, ax = plt.subplots(figsize=(10,6))
sns.histplot(np.mean(displacement_measurements,axis=1), 
             kde=False, label='Displacement')
ax.set_xlabel("Displacment [cm]",fontsize=16)
ax.set_ylabel("Frequency of occurance",fontsize=16)
plt.legend()
plt.tight_layout(), plt.xlim(2,8)
plt.savefig(fname='Distribution_AllMeasurement', dpi=150)
plt.show()
```
