```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np 

# Define Beale Function
def BealeFunc(x1, x2): 
    return ((1.5-x1+(x1*x2))**2) + ((2.25-x1+(x1*(x2**2)))**2) + ((2.625-x1+(x1*(x2**3)))**2)
def BealeFunc_minimzie():
    return ((1.5-x1+(x1*x2))**2) + ((2.25-x1+(x1*(x2**2)))**2) + ((2.625-x1+(x1*(x2**3)))**2)
# Initialize function to reset model parameters
def reset():
    x1 = tf.Variable(0.0)  
    x2 = tf.Variable(0.0)  
    return x1, x2

x1, x2 = reset()
opt = tf.keras.optimizers.Adam(learning_rate=0.02)
# Setup variable for contour plot
x1_axis = np.arange(-3.5, 3.5, 0.01) 
x2_axis = np.arange(-3.5, 3.5, 0.01) 
[X1, X2] = np.meshgrid(x1_axis, x2_axis)
Y = ((1.5-X1+(X1*X2))**2) + ((2.25-X1+(X1*(X2**2)))**2) + ((2.625-X1+(X1*(X2**3)))**2)
# Plot contour of function
plt.figure(1)
levels = [0.5, 1.0, 5.0, 10.0, 20.0, 40.0, 80.0, 500.0, 1000.0, 4000.0]
contour = plt.contour(X1, X2, Y, levels) 
plt.clabel(contour, colors = 'k', fmt = '%2.1f', fontsize=12)
plt.title(r"Contour Plot of Beale Function") 
plt.xlabel(r"$x_1$"), plt.ylabel(r"$x_2$") 
 
Z = []; Niter = []
for i in range(1000):
    print ('y = {:.1f}, x1 = {:.1f}, x2 = {:.1f}'.format(BealeFunc(x1, x2).numpy(), x1.numpy(), x2.numpy()))
    opt.minimize(BealeFunc_minimzie, var_list=[x1, x2])
    Z.append(BealeFunc(x1,x2)); Niter.append(i)
    if (i % 20 == 0):
        plt.plot(x1,x2, 'ro', markersize=4)
        plt.pause(0.2)
plt.show()

plt.figure(2)
plt.plot(Niter,Z,'b-')
plt.title('Convergence rate of Adaptive Moments (Adam) Optimizer')
plt.xlabel('Iteration'), plt.ylabel('Y value')
plt.show()
```
