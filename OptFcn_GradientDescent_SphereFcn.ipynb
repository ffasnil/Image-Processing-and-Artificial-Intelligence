```python
import tensorflow as tf
import matplotlib.pyplot as plt 
import numpy as np 

def SphereFunc(x1, x2): 
    return (x1**2.0) + (x2**2)
def SphereFunc_minimzie():
    return (x1**2.0) + (x2**2)
def reset():
    x1 = tf.Variable(10.0) 
    x2 = tf.Variable(10.0) 
    return x1, x2

x1, x2 = reset()
opt = tf.keras.optimizers.Adam(learning_rate=1.2)
# Setup variable for contour plot
x1_axis = np.arange(-15, 15, 0.2) 
x2_axis = np.arange(-15, 15, 0.1) 
[X1, X2] = np.meshgrid(x1_axis, x2_axis)
Y = np.power(X1,2) + np.power(X2,2)
# Plot contour of function
plt.figure(1)
levels = [1.0, 5.0, 10.0, 20.0, 40.0, 80.0, 120.0, 160.0, 200.0, 240.0, 280.0, 360.0]
contour = plt.contour(X1, X2, Y, levels) 
plt.clabel(contour, colors = 'k', fmt = '%2.1f', fontsize=12)
plt.title(r"Contour Plot of Sphere Function") 
plt.xlabel(r"$x_1$"), plt.ylabel(r"$x_2$") 
 
Z = []; Niter = []
for i in range(100):
    print ('y = {:.1f}, x1 = {:.1f}, x2 = {:.1f}'.format(SphereFunc(x1, x2).numpy(), x1.numpy(), x2.numpy()))
    opt.minimize(SphereFunc_minimzie, var_list=[x1,x2])
    Z.append(SphereFunc(x1,x2)); Niter.append(i)
    plt.plot(x1,x2, 'ro', markersize=5)
    plt.pause(0.2)
plt.show()

plt.figure(2)
plt.plot(Niter,Z,'b-')
plt.title('Convergence rate of Stochastic Gradient Descent (SGD) Optimizer')
plt.xlabel('Iteration'), plt.ylabel('Y value')
plt.show()
```
