```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from math import copysign, log10
```

```python
# A character ---------------------------------------------------------------------------
Aimage = cv2.imread(r'C:\Users\****\Desktop\ACharacter.png')
gray_Aimage = cv2.cvtColor(Aimage, cv2.COLOR_BGR2GRAY)
ret,Athresh = cv2.threshold(gray_Aimage,165,255,cv2.THRESH_BINARY_INV)
plt.subplot(221), plt.imshow(Athresh,'gray'), plt.title("A-Character Image")
# Calculate Moments of A character
A_moments = cv2.moments(Athresh)
# Calculate Hu Moments of A character
A_huMoments = cv2.HuMoments(A_moments)
A_loghuMoments = np.zeros((7,1))
# Hu moments in Log scale
for i in range(0,7):
    A_loghuMoments[i] = -1* copysign(1.0, A_huMoments[i]) * log10(abs(A_huMoments[i]))
    print("I_A[",i,"] = ",A_loghuMoments[i])
print("\n")
# Rotated A character ----------------------------------------------------------------------
ARimage = cv2.imread(r'C:\Users\****\Desktop\ACharacter_Mirror.png')
gray_ARimage = cv2.cvtColor(ARimage, cv2.COLOR_BGR2GRAY)
ret,ARthresh = cv2.threshold(gray_ARimage,165,255,cv2.THRESH_BINARY_INV)
plt.subplot(222), plt.imshow(ARthresh,'gray'), plt.title("Rotated A-Character Image")
# Calculate Moments of rotated A character
AR_moments = cv2.moments(ARthresh)
# Calculate Hu Moments of rotated A character
AR_huMoments = cv2.HuMoments(AR_moments)
AR_loghuMoments = np.zeros((7,1))
# Hu moments in Log scale
for i in range(0,7):
    AR_loghuMoments[i] = -1* copysign(1.0, AR_huMoments[i]) * log10(abs(AR_huMoments[i]))
    print("I_AR[",i,"] = ",AR_loghuMoments[i])
print("\n")

# E character ---------------------------------------------------------------------------    
Eimage = cv2.imread(r'C:\Users\****\Desktop\ECharacter.png')
gray_Eimage = cv2.cvtColor(Eimage, cv2.COLOR_BGR2GRAY)
ret,Ethresh = cv2.threshold(gray_Eimage,165,255,cv2.THRESH_BINARY_INV)
plt.subplot(223), plt.imshow(Ethresh,'gray'), plt.title("E-Character Image")
# Calculate Moments of E character
E_moments = cv2.moments(Ethresh)
# Calculate Hu Moments of E character
E_huMoments = cv2.HuMoments(E_moments)
E_loghuMoments = np.zeros((7,1))
# Hu moments in Log scale
for i in range(0,7):
    E_loghuMoments[i] = -1* copysign(1.0, E_huMoments[i]) * log10(abs(E_huMoments[i]))
    print("I_E[",i,"] = ",E_loghuMoments[i])
print("\n")
    
# Rotated E character ----------------------------------------------------------------------
ERimage = cv2.imread(r'C:\Users\****\Desktop\ECharacter_Mirror.png')
gray_ERimage = cv2.cvtColor(ERimage, cv2.COLOR_BGR2GRAY)
ret,ERthresh = cv2.threshold(gray_ERimage,165,255,cv2.THRESH_BINARY_INV)
plt.subplot(224), plt.imshow(ERthresh,'gray'), plt.title("Rotated E-Character Image")
# Calculate Moments of rotated E character
ER_moments = cv2.moments(ERthresh)
# Calculate Hu Moments of rotated E character
ER_huMoments = cv2.HuMoments(ER_moments)
ER_loghuMoments = np.zeros((7,1))
# Hu moments in Log scale
for i in range(0,7):
    ER_loghuMoments[i] = -1* copysign(1.0, ER_huMoments[i]) * log10(abs(ER_huMoments[i]))
    print("I_ER[",i,"] = ",ER_loghuMoments[i])
print("\n")    

# Add gap between upper and lower images
plt.subplots_adjust(hspace=0.5)

# Show images
plt.show()
```
