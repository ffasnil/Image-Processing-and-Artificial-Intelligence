```python
from sklearn.datasets import make_circles
from sklearn.decomposition import PCA
from sklearn.decomposition import KernelPCA
import matplotlib.pyplot as plt
import numpy as np

# Using scikit-learns to generate 2 concentric circles with random noise
X, y = make_circles(n_samples=1000, random_state=123, noise=0.1, factor=0.2)

# Plot Original Concentric Circle data
plt.figure(1)
plt.scatter(X[y==0, 0], X[y==0, 1], color='red', alpha=0.5)
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue', alpha=0.5)
plt.title('Concentric circles')
plt.ylabel('y coordinate'), plt.xlabel('x coordinate')

# Using Linear PCA to project concentric circle onto principle component
scikit_pca = PCA(n_components=2)
X_LinearPCA = scikit_pca.fit_transform(X)

# Plot Principle Components of Linear PCA
plt.figure(2)
plt.scatter(X_LinearPCA[y==0, 0], X_LinearPCA[y==0, 1], color='red', alpha=0.5)
plt.scatter(X_LinearPCA[y==1, 0], X_LinearPCA[y==1, 1], color='blue', alpha=0.5)
plt.ylim([-5,5])
plt.title('Principal Components after Linear PCA')
plt.xlabel('Principal Component1'), plt.ylabel('Principal Component2')

# Using Gaussian RBF kernel PCA to project concentric circle onto principle component
scikit_kpca = KernelPCA(n_components=2, kernel='rbf', gamma=5)
X_KernalPCA = scikit_kpca.fit_transform(X)

# Plot Principle Components of RBF Kernel PCA
plt.figure(3)
plt.scatter(X_KernalPCA[y==0, 0], X_KernalPCA[y==0, 1], color='red', alpha=0.5)
plt.scatter(X_KernalPCA[y==1, 0], X_KernalPCA[y==1, 1], color='blue', alpha=0.5)
plt.text(-0.05, 0.007, 'gamma = 15', fontsize=12)
plt.title('Principal Components after RBF Kernel PCA')
plt.xlabel('Principal Component1'), plt.ylabel('Principal Component2')
plt.show()
```
