```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
# params for ShiTomasi corner detection
feature_params=dict( maxCorners=100, qualityLevel=0.3, minDistance=3, blockSize=3 )

# Reading an image in default mode  
frame1 = cv2.imread(r'C:\Users\ffah\Desktop\Projects\BEng Project\SIIT ML\Robotics Lecture\Robotics_code\week5\Figures\CarOriginal.jpg')
frame2 = cv2.imread(r'C:\Users\ffah\Desktop\Projects\BEng Project\SIIT ML\Robotics Lecture\Robotics_code\week5\Figures\CarCloser.jpg')
# "cvtColor" : convert color- to gray-image
gray_f1 = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
gray_f2 = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)

plt.subplot(221), plt.imshow(gray_f1,cmap = 'gray'), plt.title("Grayscale Frame1")
plt.subplot(222), plt.imshow(gray_f2,cmap = 'gray'), plt.title("Grayscale Frame2")

# Finds the strongest corners in 1st and 2nd frame by Shi-Tomasi method - optical flow can use these corners
ptcorner1 = cv2.goodFeaturesToTrack(gray_f1, mask = None, **feature_params)
ptcorner2 = cv2.goodFeaturesToTrack(gray_f2, mask = None, **feature_params)

ptcorner1 = np.intp(ptcorner1) # "int0" is an alias for intp
ptcorner2 = np.intp(ptcorner2)

for i in ptcorner1:
    x,y = i.ravel() # flattened high-dimensional array into one-dimensional array. 
    cv2.circle(frame1,(x,y),5,(0,0,255),-1)
for i in ptcorner2:
    x,y = i.ravel() # flattened high-dimensional array into one-dimensional array. 
    cv2.circle(frame2,(x,y),5,(0,0,255),-1)

plt.subplot(223), plt.imshow(cv2.cvtColor(frame1, cv2.COLOR_BGR2RGB)), plt.title("Corner points in Frame1")
plt.subplot(224), plt.imshow(cv2.cvtColor(frame2, cv2.COLOR_BGR2RGB)), plt.title("Corner points in Frame2")
plt.show()
```
